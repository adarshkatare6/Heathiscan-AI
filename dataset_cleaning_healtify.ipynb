```python
import pandas as pd
impot numpy as np
# load dataset
df = pd.read_csv("openfood_data.csv")

# keep only required columns
df = df[["product_name", "ingredients_text","serving_size"]]

# remove null values
df = df.dropna(subset=["product_name", "ingredients_text","serving_size"])

# count words in ingredients column
df["word_count"] = df["ingredients_text"].astype(str).apply(lambda x: len(x.split()))

# keep rows with 5–20 words
df = df[(df["word_count"] >= 5) & (df["word_count"] <= 20)]

# remove helper column
df = df.drop(columns=["word_count"])

# reset index
df = df.reset_index(drop=True)

# save cleaned dataset
df.to_csv("final_dataset.csv", index=False)

print("Final dataset size:", df.shape)
```
